```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting visualization style
sns.set(style="whitegrid")
```

```python
ttc_2022 = pd.read_csv('ttc-bus-delay-data-2022.csv')
```

```python
ttc_2023 = pd.read_excel('ttc-bus-delay-data-2023.xlsx')
```

```python
print("2022 Dataset:")
print(ttc_2022.head())
print("\n2023 Dataset:")
print(ttc_2023.head())
```

```python
print("2022 Dataset Info:")
print(ttc_2022.info())
print("\n2023 Dataset Info:")
print(ttc_2023.info())
```

```python
print("\nMissing values in 2022 data:")
print(ttc_2022.isnull().sum())
print("\nMissing values in 2023 data:")
print(ttc_2023.isnull().sum())
```

```python
# Handling missing values in Route column
ttc_2022['Route'].fillna('Unknown')
ttc_2023['Route'].fillna('Unknown')
```

```python
# Handling missing values in Direction column
ttc_2022['Direction'].fillna('Unknown', inplace=True)
ttc_2023['Direction'].fillna('Unknown', inplace=True)
```

```python
# Verifying that the missing values have been handled
print("Missing values after handling:")
print("2022 Dataset:")
print(ttc_2022.isnull().sum())
print("\n2023 Dataset:")
print(ttc_2023.isnull().sum())
```

```python
# Adding a column to distinguish years
ttc_2022['year'] = 2022
ttc_2023['year'] = 2023
```

```python
# Merging the datasets
ttc_data = pd.concat([ttc_2022, ttc_2023], ignore_index=True)
```

```python
# Verifying the merged dataset
print("Merged Dataset:")
print(ttc_data.head())
```

```python
# Fixing the date format in the merged dataset
from dateutil import parser

def convert_date(Date):
    try:
        # Try parsing using the 2022 format first
        return pd.to_datetime(Date, format="%d-%b-%y", errors='coerce')
    except:
        # Fallback to automatic parsing (handles 2023 format)
        return pd.to_datetime(Date, errors='coerce')

# Applying the function to the 'date' column
ttc_data['Date'] = ttc_data['Date'].apply(convert_date)

# Verifying the conversion
print("Merged Dataset - Date Column After Conversion:")
print(ttc_data['Date'].head())
```

```python
print("Summary Statistics:")
print(ttc_data.describe())
```

```python
print("\nUnique Routes:")
print(ttc_data['Route'].nunique())
```

```python
print("Unique Incidents:")
print(ttc_data['Incident'].nunique())
```

```python
print("Unique Directions:")
print(ttc_data['Direction'].nunique())
```

```python
print("\nBasic Information:")
print(ttc_data.info())
```

```python
# Extracting hour from the time column
ttc_data['hour'] = pd.to_datetime(ttc_data['Time'], errors='coerce').dt.hour

# Average delay per hour
avg_delay_hour = ttc_data.groupby('hour')['Min Delay'].mean()

plt.figure(figsize=(10, 6))
sns.lineplot(x=avg_delay_hour.index, y=avg_delay_hour.values, marker='o', color='green')
plt.title("Average Delay per Hour")
plt.xlabel("Hour")
plt.ylabel("Average Delay (Minutes)")
plt.show()
```

```python
# Categorizing delays as short, medium, long
ttc_data['delay_category'] = pd.cut(ttc_data['Min Delay'], bins=[0, 5, 15, 60, 180], labels=['Short', 'Medium', 'Long', 'Very Long'])

# Distribution of delay categories
plt.figure(figsize=(8, 5))
sns.countplot(data=ttc_data, x='delay_category', palette='pastel')
plt.title("Distribution of Delay Categories")
plt.xlabel("Delay Category")
plt.ylabel("Count")
plt.show()
```

```python
# Creating a time of day category
def categorize_time(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

# Adding a 'time_of_day' column
ttc_data['hour'] = pd.to_datetime(ttc_data['Time'], errors='coerce').dt.hour
ttc_data['time_of_day'] = ttc_data['hour'].apply(categorize_time)

# Count of delays by time of day
time_of_day_count = ttc_data['time_of_day'].value_counts()

# Plotting the count of delays by time of day
plt.figure(figsize=(8, 5))
sns.barplot(x=time_of_day_count.index, y=time_of_day_count.values, palette='Set2')
plt.title("Number of Delays by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Number of Delays")
plt.show()
```

```python
# Average delay by incident type
avg_delay_incident = ttc_data.groupby('Incident')['Min Delay'].mean().sort_values(ascending=False).head(10)

# Plotting average delay by incident type
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_delay_incident.values, y=avg_delay_incident.index, palette='Spectral')
plt.title("Average Delay by Incident Type (Top 10)")
plt.xlabel("Average Delay (Minutes)")
plt.ylabel("Incident Type")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Selecting features for clustering
features = ttc_data[['Min Delay', 'Min Gap', 'hour']].dropna()

# Standardizing the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Applying K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(scaled_features)

# Adding cluster labels to the data
ttc_data['cluster'] = -1
ttc_data.loc[features.index, 'cluster'] = clusters

# Plotting clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=ttc_data, x='hour', y='Min Delay', hue='cluster', palette='tab10', alpha=0.7)
plt.title("Delay Clusters Based on Time and Duration")
plt.xlabel("Hour of the Day")
plt.ylabel("Delay (Minutes)")
plt.show()
```

```python
# Analyzing cluster characteristics
cluster_summary = ttc_data.groupby('cluster').agg({
    'Min Delay': ['mean', 'median'],
    'hour': ['mean', 'median'],
    'Route': 'nunique',
    'Incident': 'nunique'
})

print("Cluster Summary:")
print(cluster_summary)
```
